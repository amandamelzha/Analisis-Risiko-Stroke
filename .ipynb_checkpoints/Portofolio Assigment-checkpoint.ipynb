# ABOUT DATASET

| Variabel             | Keterangan                                                                                                                                                                                                             |
| --------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **id**                | Nomor identitas unik untuk setiap pasien (tidak digunakan dalam analisis).                                                                                                                                             |
| **gender**            | Jenis kelamin pasien (Male, Female, Other).                                                                                                                                                                            |
| **age**               | Usia pasien (dalam tahun).                                                                                                                                                                                             |
| **hypertension**      | Status hipertensi pasien: 1 = memiliki hipertensi, 0 = tidak memiliki hipertensi.                                                                                                                                      |
| **heart_disease**     | Status penyakit jantung: 1 = memiliki penyakit jantung, 0 = tidak memiliki penyakit jantung.                                                                                                                           |
| **ever_married**      | Status pernikahan: "Yes" = pernah/sedang menikah, "No" = belum pernah menikah.                                                                                                                                         |
| **work_type**         | Jenis pekerjaan pasien: <br> • private = karyawan swasta <br> • self-employed = wiraswasta <br> • govt_job = pegawai pemerintah <br> • never_worked = tidak pernah bekerja <br> • children = anak-anak (belum bekerja) |
| **Residence_type**    | Jenis tempat tinggal: <br> • Urban = daerah perkotaan <br> • Rural = daerah pedesaan                                                                                                                                   |
| **avg_glucose_level** | Rata-rata kadar glukosa darah pasien.                                                                                                                                                                                  |
| **bmi**               | Indeks Massa Tubuh (Body Mass Index).                                                                                                                                                                                  |
| **smoking_status**    | Status merokok: <br> • never smoked = tidak pernah merokok <br> • formerly smoked = pernah merokok (sudah berhenti) <br> • smokes = masih merokok <br> • unknown = tidak diketahui                                     |
| **stroke**            | Status stroke: 1 = pernah mengalami stroke, 0 = tidak pernah mengalami stroke.                                                                                                                                         |

Distribusi usia : menunjukkan dominasi populasi dewasa hingga lansia. Nilai median yang sedikit lebih tinggi dibandingkan mean mengindikasikan distribusi yang sedikit left-skewed (negatively skewed).
Distribusi Kada Gula : Nilai mean yang jauh lebih tinggi dibandingkan median menunjukkan bahwa distribusi kadar glukosa darah bersifat right-skewed.Hal ini mengindikasikan adanya sejumlah pasien dengan kadar glukosa yang sangat tinggi.
Distribusi BMI: Nilai mean dan median yang hampir sama menunjukkan bahwa distribusi BMI relatif mendekati normal (simetris).

```python
# Menghitung Q1, Q3, dan IQR untuk kolom Age
Q1 = df["age"].quantile(0.25)
Q3 = df["age"].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengecek jumlah outlier
outlier_age = df[
    (df["age"] < lower_bound) | 
    (df["age"] > upper_bound)
]

# Menampilkan Output
print("Batas bawah:", lower_bound)
print("Batas atas :", upper_bound)
print("Jumlah Outlier Age:", len(outlier_age))
```

```python
# Menghitung Q1, Q3, dan IQR untuk kolom avg_glucose_level
Q1 = df["avg_glucose_level"].quantile(0.25)
Q3 = df["avg_glucose_level"].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengecek jumlah outlier
outlier_glucose = df[
    (df["avg_glucose_level"] < lower_bound) | 
    (df["avg_glucose_level"] > upper_bound)
]

# Menampilkan Output
print("Batas bawah:", lower_bound)
print("Batas atas :", upper_bound)
print("Jumlah Outlier avg_glucose_level:", len(outlier_glucose))
```

```python
# Menghitung Q1, Q3, dan IQR untuk kolom BMI
Q1 = df["bmi"].quantile(0.25)
Q3 = df["bmi"].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengecek jumlah outlier
outlier_BMI = df[
    (df["bmi"] < lower_bound) | 
    (df["bmi"] > upper_bound)
]

# Menampilkan Output
print("Batas bawah:", lower_bound)
print("Batas atas :", upper_bound)
print("Jumlah Outlier BMI:", len(outlier_BMI))
```

```python
# Membuat kategori BMI 
df_clean['bmi_category'] = df_clean['bmi'].apply(lambda x: 'Underweight' if x <= 18.49 else ('Normal' if 18.5 <= x <= 24.9 else ('Overweight' if 25 <= x <= 29.9 else 'Obese')))

# Lihat hasil beberapa baris
df_clean[['bmi', 'bmi_category']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
sns.set_style("white")

fig, axes = plt.subplots(1, 2, figsize=(12,4))

# Usia
sns.kdeplot(data=df_clean,
            x='age',
            hue='stroke',
            fill=True,
            common_norm=False,
            alpha=0.4,
            palette=['#A8D0E6', '#F4978E'],
            ax=axes[0])

axes[0].set_title("Distribusi Usia berdasarkan Stroke", fontsize=12)
axes[0].set_xlabel("Usia")
axes[0].set_ylabel("Density")
axes[0].grid(False)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

# Glukosa
sns.kdeplot(data=df_clean,
            x='avg_glucose_level',
            hue='stroke',
            fill=True,
            common_norm=False,
            alpha=0.4,
            palette=['#A8D0E6', '#F4978E'],
            ax=axes[1])

axes[1].set_title("Distribusi Glukosa berdasarkan Stroke", fontsize=12)
axes[1].set_xlabel("Kadar Glukosa")
axes[1].set_ylabel("")
axes[1].grid(False)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

Distribusi usia menunjukkan bahwa pasien stroke cenderung berada pada kelompok usia yang lebih tua. Kurva kepadatan stroke bergeser ke kanan, menandakan peningkatan risiko pada usia lanjut.
Distribusi kadar glukosa menunjukkan bahwa pasien dengan glukosa tinggi memiliki kecenderungan lebih besar mengalami stroke, terlihat dari kepadatan yang lebih tinggi pada nilai glukosa di atas 150.


```python
import pandas as pd

plt.style.use('default')
sns.set_style("white")

# Hitung proporsi
hyper_stroke = pd.crosstab(df_clean['hypertension'],
                           df_clean['stroke'],
                           normalize='index') * 100

heart_stroke = pd.crosstab(df_clean['heart_disease'],
                           df_clean['stroke'],
                           normalize='index') * 100

hyper_stroke.index = ['Tidak Hipertensi', 'Hipertensi']
heart_stroke.index = ['Tidak Penyakit Jantung', 'Penyakit Jantung']

fig, axes = plt.subplots(1, 2, figsize=(12,4))

#  Hipertensi
hyper_stroke.plot(kind='barh',
                  stacked=True,
                  color=['#A8D0E6', '#F4978E'],
                  ax=axes[0],
                  legend=False)

axes[0].set_xlim(0,100)
axes[0].set_title("Stroke vs Hipertensi", fontsize=12)
axes[0].set_xlabel("Persentase (%)")
axes[0].set_ylabel("")
axes[0].grid(False)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.1f%%', label_type='center', fontsize=8)


# Heart Disease
heart_stroke.plot(kind='barh',
                  stacked=True,
                  color=['#A8D0E6', '#F4978E'],
                  ax=axes[1],
                  legend=False)

axes[1].set_xlim(0,100)
axes[1].set_title("Stroke vs Penyakit Jantung", fontsize=12)
axes[1].set_xlabel("Persentase (%)")
axes[1].set_ylabel("")
axes[1].grid(False)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.1f%%', label_type='center', fontsize=8)

fig.legend(["Tidak Stroke", "Stroke"],
           loc='upper center',
           ncol=2,
           frameon=False)

plt.tight_layout()
plt.show()
```

Analisis menunjukkan bahwa pasien dengan hipertensi memiliki proporsi stroke yang jauh lebih tinggi dibandingkan pasien tanpa hipertensi. 
Selain itu, pasien dengan penyakit jantung menunjukkan proporsi stroke yang lebih tinggi lagi, menjadikannya faktor risiko paling dominan dalam dataset ini.
Temuan ini mengindikasikan bahwa kondisi seperti hipertensi dan penyakit jantung berperan signifikan dalam peningkatan risiko stroke.

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('default')

#  BMI 
bmi_order = ['Underweight', 'Normal', 'Overweight', 'Obese']

# Hitung proporsi
bmi_stroke = pd.crosstab(df_clean['bmi_category'],
                         df_clean['stroke'],
                         normalize='index') * 100

# Urutkan sesuai kategori logis
bmi_stroke = bmi_stroke.reindex(bmi_order)

res_stroke = pd.crosstab(df_clean['Residence_type'],
                         df_clean['stroke'],
                         normalize='index') * 100

fig, axes = plt.subplots(1, 2, figsize=(14,6))

# BMI Category 

bmi_stroke.plot(kind='barh',
                stacked=True,
                color=['#A8D0E6', '#F4978E'],
                ax=axes[0],
                legend=False)

axes[0].set_xlim(0,100)
axes[0].set_title("Stroke vs Kategori BMI", fontsize=13)
axes[0].set_xlabel("Persentase (%)")
axes[0].set_ylabel("")
axes[0].invert_yaxis()  # supaya urutan dari bawah ke atas lebih natural
axes[0].grid(False)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)

for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.1f%%', label_type='center', fontsize=9)

# Residence Type (Horizontal)
res_stroke.plot(kind='barh',
                stacked=True,
                color=['#A8D0E6', '#F4978E'],
                ax=axes[1],
                legend=False)

axes[1].set_xlim(0,100)
axes[1].set_title("Stroke vs Residence Type", fontsize=13)
axes[1].set_xlabel("Persentase (%)")
axes[1].set_ylabel("")
axes[1].grid(False)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.1f%%', label_type='center', fontsize=9)

# Legend bersama
fig.legend(["Tidak Stroke", "Stroke"],
           loc='upper center',
           ncol=2,
           frameon=False)

plt.subplots_adjust(wspace=0.35)
plt.tight_layout()
plt.show()
```

Proporsi stroke cenderung meningkat pada kategori BMI Overweight dan Obese dibandingkan kategori Normal.
Namun, perbedaannya tidak setajam faktor hipertensi dan penyakit jantung.
Proporsi stroke antara area Urban dan Rural relatif serupa, sehingga faktor lokasi tempat tinggal tampak tidak memiliki pengaruh kuat terhadap kejadian stroke dalam dataset ini.
